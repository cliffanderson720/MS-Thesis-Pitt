```python
# I'm trying to implement BCs and Godunov at the same time. I have a few options:
#   1- duplicate Qiqi example
#   2- change current setup to integrate with Godunov
#   3- do 2, and incorporate second BC.

# doing 2
# Qiqi sets cell values to 0 for BCs, then makes cell edges.
# ρ and q functions should output an array of cell fluxes with BCs.
# Godunov function can make that into interfaces

def get_ρ(ρ): 
#     ρ[0] = 0
#     ρ[-1] = 0
    return ρ

def get_c(ρ):
    c = np.ones(len(ρ))
    c[np.where(ρ>=0.95)] = 0.95
    v = (1-ρ)**2
    return v

def get_q(ρ):
    return  get_c(ρ) * (get_ρ(ρ))

# def rhsf(ρ,t):
#     qf = get_qf(ρ)
#     dρdt = -1/Δx*(qf[1:]-qf[:-1])
#     return dρdt

def godunov(ρ,t):
    ρ_bc = np.zeros(ngrd+2)
    ρ_bc[-1] = 1.0
    ρ_bc[1:-1] = ρ

    q = get_q(ρ_bc)
    
    # evaluate fluxes at each cell wall 
    qf = np.zeros(ngrdf)
    for i in range(ngrdf):
        if ρ_bc[i] >= ρ_bc[i+1]:
            qf[i] = max(q[i],q[i+1])
        elif ρ_bc[i] <= 0 and 0 <= ρ_bc[i+1]:
            qf[i] = 0
        else:
            qf[i] = min(q[i],q[i+1])
            
    dρdt = 1/Δx*(qf[:-1]-qf[1:])
    return dρdt
        

#---------------------------

ngrd = 250
L    = 1
tend = 50
ntimes = 500

times = np.linspace(0,tend,ntimes)
ngrdf = ngrd+1
Δx    = L/ngrd
xf    = np.linspace(0,L,ngrdf)
x     = (xf[1:] + xf[:-1])/2

fig,axs = plt.subplots(2)
fig.figsize=(10,10)

for i,init in enumerate((0.4,0.6)):
    ρ0  = np.ones(ngrd)*init
    ρall = odeint(godunov, ρ0, times)


    color = plt.get_cmap('Greys')
    colors = iter(color(np.linspace(0.0,1,ntimes)))
    for j in range(ntimes):
        axs[i].plot(x,ρall[j,:],color=next(colors))
        axs[i].set_ylabel(str(init))
        axs[i].set_xlim((0,L))

plt.xlabel('position')
# plt.ylabel(r'$\rho$')
plt.xlim([0,L]);
plt.ylim([0,1]);
# plt.legend(loc=0);
plt.show()
```
