```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
from scipy.integrate import odeint
np.set_printoptions(precision=3)
from itertools import cycle
```

#### Goals:
1. Write rhsf to propagate shock either direction based on sign of speed
2. Extend rhsf to 2 waves

```python
def get_ρb(ρ,source=ρ0[0]): 
    '''applies 2 boundary cells on the ends of the ρ array'''
    ρb = np.zeros(ngrd+2)
    ρb[1:-1] = ρ[:]
    ρb[-1] = ρ[-1]
    ρb[0]  = ρ[0]
    return ρb

def get_c(ρ):
    '''computes a wave speed based on concentration'''
    c = (ρ)**1
    return c

def rhsf(ρ,t):
    '''returns derivative'''
    if t<0.5:
        source=0.4
    else:
        source = ρ0[0]
    ρbc = get_ρb(ρ,source)
    c   = get_c(ρ)
    
    # split speeds based on sign
    i_pos        = np.where(c>=0)[0]
    cpos,cneg    = [np.zeros_like(c) for i in range(2)]
    cpos[i_pos]  = c[i_pos]
    cneg[~i_pos] = c[~i_pos]
    
    # designate Qi-1, Qi, and Qi+1
    Ql = ρbc[0:-2]
    Qi = ρbc[1:-1]
    Qr = ρbc[2:]
    dρdt = -1/Δx*(cpos*(Qi-Ql) + cneg*(Qr-Qi))
    
    return dρdt#,c,cneg,cpos,ρbc,Ql,Qi,Qr

ngrd = 200
L    = 5
tend = 2.5 # higher than this, the solution is unstable
ntimes = 50

times = np.linspace(0,tend,ntimes)
ngrdf = ngrd+1
Δx    = L/ngrd
xf    = np.linspace(0,L,ngrdf)
x     = (xf[1:] + xf[:-1])/2

ρ0  = np.linspace(1,-1,ngrd)
rhsf(ρ0,1)

ρall = odeint(rhsf, ρ0, times)

blue = iter(plt.cm.Blues(np.linspace(0.2,1,ntimes)))
for j in range(ntimes):
    plt.plot(x,ρall[j,:],label=round(times[j],1),c=next(blue))
    
plt.xlabel('position')
plt.ylabel(r'$\rho$')
plt.xlim([0,L]);
# plt.legend(loc=0)
plt.show()
```
