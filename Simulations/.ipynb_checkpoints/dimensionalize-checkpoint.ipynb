#### Dimensionalize simulations with variable material properties
I have nondimensional time for RBC sedimentation. I need to convert that to dimensional time, but the result depends on the physical characteristics of the blood. Write functions to get dimensional time from variable material properties and probe parameter sensitivity.  

Support experiment setup by calculating dilution hematocrits

```python
from scipy.optimize import minimize,fsolve
from scipy.misc import derivative
import numpy as np
import velocity_functions as vf
import matplotlib.pyplot as plt
```

#### Convert from non-dimensional time to dimensioned time

```python
def dimtime(nondim_time,L,u1_inf):
    return  nondim_time*L/u1_inf

nL = 51
nt = 51
nd = 101
nr = 101

myndt = 2.04 # dimensionless time for H=40
myr   = 2.6e-6
myρ   = 1110

ρlo,ρhi = 1085,1120
rlo,rhi = 2.5e-6,5e-6

L0 = 0.0565
Lw = 0.06
Ls = np.linspace(L0,Lw,nL) # m
density = np.linspace(ρlo,ρhi,nd)
radius  = np.linspace(rlo,rhi,nr)
nondim_time = np.linspace(1,2,nt)

uρlo,uρhi = vf.stokes(3000,Lw,myr,ρlo),vf.stokes(3000,Lw,myr,ρhi)
urlo,urhi = vf.stokes(3000,Lw,rlo,myρ),  vf.stokes(3000,Lw,rhi,myρ)

myu0  = vf.stokes(3000,Lw,myr,myρ)
mytf  = dimtime(myndt,Lw-L0,myu0)

tρlo,tρhi = dimtime(myndt,Lw-L0,uρlo),dimtime(myndt,Lw-L0,uρhi)
trlo,trhi = dimtime(myndt,Lw-L0,urlo),dimtime(myndt,Lw-L0,urhi)

# Variability in stokes velocity and dimensional time with cell parameters
u0 = np.empty((nr,nd))
tf = np.empty((nr,nd))
for r in range(nr):
    for d in range(nd):
        u0[r,d] = vf.stokes(3000,Lw,radius[r],density[d],visc=0.0013)
        tf[r,d] = dimtime(myndt,Lw-L0,u0[r,d])
```

Sedimentation time for H=0.25

```python
fig,ax = plt.subplots(ncols=2,figsize=(12,4),sharex=True) #,gridspec_kw={'width_ratios': [1, 1]})
levels = 101
D,R    = np.meshgrid(density,radius*1e6)
u0plot = ax[0].contourf(R,D,u0*1e3,levels,cmap=plt.cm.jet)
tfplot = ax[1].contourf(R,D,tf,    levels,cmap=plt.cm.jet)

fig.colorbar(u0plot,ax=ax[0],label=f'$u_\infty$ mm/s')
fig.colorbar(tfplot,ax=ax[1],label=f'Time (s) to RBC sedimentation')

ax[0].text(myr*1e6+.05,myρ,f'{myu0*1e3:.3f}')
ax[1].text(myr*1e6+.05,myρ,f'{mytf:.0f}')
[ax[0].text(r*1e6,ρ,f'{u*1e3:.3f}') for r,ρ,u in zip((rlo,rhi,myr,myr),(myρ,myρ,ρlo,ρhi),(urlo,urhi,uρlo,uρhi))]
[ax[1].text(r*1e6,ρ,f'{t:.0f}') for r,ρ,t in zip((rlo,rhi,myr,myr),(myρ,myρ,ρlo,ρhi),(trlo,trhi,tρlo,tρhi))]

[axx.set_xlabel(f'Radius ($\mu m$)') for axx in ax]
ax[0].set_ylabel(f'Density  ($kg/m^3$)')
plt.show()
print(f'Final time for ρ {ρlo},{ρhi}   is {tρlo:.0f},{tρhi:.0f} s')
print(f'Final time for r {rlo},{rhi} is {trlo:.0f},{trhi:.0f} s')
```

```python
# Find hematocrits to optimize flux
def michaels_flux(ϕ,**kwargs):
    return vf.michaels(ϕ,2.71,amax=0.95)*ϕ*(1-ϕ)

ϕ_fm = minimize(lambda ϕ: -michaels_flux(ϕ,n=2.71,amax=0.95),0.2) # density for max flux
ϕ_lo = 0.05
ϕ_hi = fsolve(lambda ϕ_hi: michaels_flux(ϕ_hi)-michaels_flux(ϕ_lo),0.4)[0]
ϕlevels = np.array([ϕ_lo,ϕ_fm.x[0],ϕ_hi])

# Find leading rarefaction ϕ from a given hematocrit
# get derivative of michaels flux. Nonlinear solve for where it intercepts the flux curve
def dfdϕ(ϕ,n=2.71,ϕmax=0.95):
    '''Derivative of michaels bolger flux function. Derived by hand 3 May 2019'''
    
    return (1-ϕ)**3*ϕ*(-n*(1-ϕ/ϕmax)**(n*ϕmax-1)) + \
           (1-ϕ/ϕmax)**(n*ϕmax)*((1-ϕ)**3 - ϕ*3*(1-ϕ)**2)

def rare_curve(ϕ0,curve):
    '''
    get rarefaction ϕ given ϕ0 by solving a tangency problem.
    Returns ϕ0 upon failure of any element'''
    ϕ_inflection = fsolve(lambda ϕ_inflect: derivative(michaels_flux,ϕ_inflect,n=2,dx=1e-6),0.3)[0]
    
    if (ϕ0>=ϕ_inflection).any():
        raise Exception(f'ϕ cannot exceed ϕ_inflection ({ϕ_inflection:.4f})')
        return ϕ0
    
    def tangency(ϕ_rare):
        return curve(ϕ_rare)-(michaels_flux(ϕ_rare)-michaels_flux(ϕ0))/(ϕ_rare-ϕ0)
    
    ϕ_rare_guess = np.ones_like(ϕ0)*0.3
    return fsolve(tangency,ϕ_rare_guess)

    
# ϕ_rare = rare_curve(np.array(ϕlevels),dfdϕ)
# ϕ_rare_pair = np.vstack([ϕlevels,ϕ_rare])

amax = 0.95
n = 101 
ϕs = np.linspace(0,amax,n)
plt.plot(ϕs*100,michaels_flux(ϕs))
# plt.plot(ϕ_rare_pair,michaels_flux(ϕ_rare_pair),'o--')
plt.scatter(ϕlevels*100,michaels_flux(ϕlevels))

plt.xlabel('Hematocrit (%)')
plt.ylabel('Red cell flux (a.u.)' )
plt.grid()

print(f'ϕ,f for maximum flux:     {ϕ_fm.x[0]:0.3f},       {-ϕ_fm.fun:0.3f}')
print(f'ϕ,f for equal fluxes:     {ϕ_lo:.3f} {ϕ_hi:.3f}, {michaels_flux(ϕ_hi,n=2.71,amax=0.95):.3f}')
```

```python
def parallel(ϕstar,ϕl,ϕr):
    return dfdϕ(ϕstar)-(michaels_flux(ϕr)-michaels_flux(ϕl))/(ϕr-ϕl)
fsolve(parallel,0.2,args=(.15,0.3))
```

```python
# find hematocrits to maximize backflow velocity
def vplas(ϕ):
    return -ϕ*vf.michaels(ϕ)

# hematocrit for max backflow velocity
ϕ_vm = minimize(vplas,0.2) 

# get hematocrits for half of max velocity. Use values higher and lower than predicted best.
ϕ_halfv_lo,ϕ_halfv_hi = [fsolve(lambda ϕ: vplas(ϕ_vm.x[0])/2 - vplas(ϕ),guess)[0] for guess in (0.05,0.4)]

# combine into one array
ϕvlevels = np.array([ϕ_halfv_lo,ϕ_vm.x[0],ϕ_halfv_hi])

plt.plot(ϕs*100,vplas(ϕs))
plt.scatter(ϕvlevels*100,vplas(ϕvlevels))
print(f'ϕ,v for maximum velocity: {ϕ_vm.x[0]:0.3f},       {ϕ_vm.fun:0.3f}')
print(f'ϕ,v for equal velocities: {ϕ_halfv_lo:.3f} {ϕ_halfv_hi:.3f}, {vplas(ϕ_halfv_lo):.3f}')
```
