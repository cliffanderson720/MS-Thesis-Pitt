```python
def oneD_to_bc(oneDarrays,bcs):
    '''Appends boundary cells to 1D arrays on concentration.
    Wrapper to ODE-type functions where odeint requires a 1d array.
    --------------------------------------------------------------
    Inputs:
        oneDarrays: tuple of 1d (n,) arrays
        bcs: tuple of scalar boundary conditions. If boundary values
             are heterogeneous, give a tuple of tuples of boundary values
             e.g.: (0,1) for homogeneous, ((0,1),(2,3)) for [0,a,1,2,b,3]
    --------------------------------------------------------------
    Output: flattened array of concentrations with ghost cells'''
    
    twoD = np.vstack(oneDarrays)
    twoDbc = np.insert(twoD,[0,twoD.shape[1]],bcs,axis=1)
    return twoDbc.flatten()

myarr = oneD_to_bc((ρ10,ρ20),((0,1),(2,4)))
print(myarr)
i = 0
nrgd = 5
myarr[i*(nrgd+2):(i+1)*(ngrd+2)]
```

```python
def godunov2wave(ρ_bc,t,printmode=False):
    # Solve for the interface fluxes one section at a time.
    # Store fluxes in output
    
    q = -(1-ρ_bc)**2

    # evaluate fluxes at each cell wall 
    qf = np.zeros(ngrdf*2)
    if printmode==True: print('cond, ρ[i], ρ[i+1], qi, q[i+1]  qf')
    for i in range((ngrd+1)*2-1):
        if ρ_bc[i] >= ρ_bc[i+1]:
            qf[i] = max(q[i],q[i+1])
            cond = 1
            
        elif ρ_bc[i] <= 0 and 0 <= ρ_bc[i+1]:
            qf[i] = 0
            cond=2
            
        else:
            qf[i] = min(q[i],q[i+1])
            cond=3
    
        if printmode==True:
            print('   {0}  {1:.2f}  {2:.2f}   {3:.2f} {4:.2f} {5:.3f}'.format(cond,ρ_bc[i],ρ_bc[i+1],q[i],q[i+1],qf[i]))
    if printmode==True: print()

    dρdt = 1/Δx*(qf[:-1]-qf[1:])
    return dρdt

ngrd = 5
L    = 5
tend = 1
ntimes = 4

times = np.linspace(0,tend,ntimes)
ngrdf = ngrd+1
Δx    = L/ngrd
xf    = np.linspace(0,L,ngrdf)
x     = (xf[1:] + xf[:-1])/2

ρ10  = np.ones(ngrd)*0.5
ρ20  = np.ones(ngrd)*0.3
ρ0 = oneD_to_bc((ρ10,ρ20),(0,1))
ρall = odeint(godunov2wave, ρ0, times)
```

```python
# %matplotlib notebook
color = plt.get_cmap('Reds')
colors = iter(color(np.linspace(0.2,1,ntimes)))
for j in range(ntimes):
    plt.plot(x,ρall[j,:],color=next(colors))
    print(godunov(ρall[j,:],1,True))

plt.xlabel('Normalized position')
plt.ylabel(r'$Hematocrit$')
plt.xlim([0,L]);
```

```python
def rk4step(rhsf, y):               # 4th order Runge-Kutta solver
    
    Δy1 = Δt*rhsf(y,           Δt)[0]
    Δy2 = Δt*rhsf(y + 0.5*Δy1, Δt/2)[0]
    Δy3 = Δt*rhsf(y + 0.5*Δy2, Δt/2)[0]
    Δy4 = Δt*rhsf(y +     Δy3, Δt)[0]
    
    return y + 1/6*(Δy1 + 2*Δy2 + 2*Δy3 + Δy4)
#---------------------------
```
